## Selection and Filtering with .loc and .iloc

```python
# import libraries
import pandas as pd
import numpy as np
from numpy import random

from IPython import display
```

### Create a dataframe for ten cities with average temperatures for each month

```python
# create a 10 x 12 of random temperatures
temps = random.randint(-10,105,size=120).reshape(10,12)
```

```python
# create row and column labels
cities =['Auckland','Simla','Bern','Taganrog','Jakarta','Funafuti','Hanoi','Kiev','Moroni','Quito']
months =['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
print(type(temps))
print(type(cities))
print(type(months))
```

```python
# create dataframe
df = pd.DataFrame(data=temps,index=cities,columns=months)
```

## Referring to Rows and Columns

```python
# access a row by explicit index or label; returns a series
df.loc['Bern']
```

```python
# returns a dataframe
df.loc[['Bern']]
```

```python
# accessing a row by implicit index or integer position; again returns a series
df.iloc[2]
```

```python
# selecting a column by label
df.loc[:,'May']
```

```python
# selecting a column by column position
df.iloc[:,4]
```

```python
# selecting rows and columns by row and column position
df.iloc[3:6,2:6]
```

```python
# selecting rows and columns by row and column position
df.loc["Taganrog":"Funafuti","Mar":"Jun"]
```

```python
# filtering according to criterion and returning only certain columns
df.loc[df.Jan>50,'Jun':'Sep']
```

```python
# returns all columns
df.loc[df.Jan>50]
```

```python
# filtering and returning only specified columns
df.loc[df.Jan>50,:][['Jul','Sep']]
```
